```python
# entrada
lado1 = float (input("digite o comprimento:"))
lado2 = float (input("digite a altura:"))

# processamento

area = lado1 * lado2

# saida

print (f"O total da area é :{area} metros quadrados")
```

```python
# entrada
numero1 = float (input("digite o numero :"))

# processamento

dobro = numero1 * 2

# saida
print (f"O dobro de numero é :{dobro}")
```

```python
# entrada
nota1 = float (input("digite a 1º nota :"))
nota2 =  float (input("digite a 2º nota :"))
nota3 = float (input("digite a 3º nota :"))
nota4 = float (input("digite a 4º nota :"))

# processamento

media = nota1 + nota2 + nota3 + nota4
media2 = media / 4

# saida

print (f"O resultado da media das notas é :{media2}")
```
