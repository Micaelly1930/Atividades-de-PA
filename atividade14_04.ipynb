```python
# Par ou Ímpar

# Entrada
numero = int(input("Digite um número: "))

# Processamento e Saída
if numero % 2 == 0:
    print(f"O número {numero} é par.")
else:
    print(f"O número {numero} é ímpar.")
```

```python
# positivo, negativo ou zero

#entrada

# < que 0 negativo
# > que 0 positivo
# == 0 zero

numero = int(input(" digite um numero"))


#procesamento e saida

if numero < 0:
  print(f" esse numero e negativo")
elif numero > 0:
  print(f" esse numero e positivo")
else:
  print(f" esse numero e zero")
```

```python
# calculo

#pedir para o usuario escrever 2 numeros

#entrada

a = int(input("digite um numero: "))
b = int(input("digite um numero"))

#procesamento

soma = a + b
subtraçao = a - b
mutiplicaçao = a * b
divisao = a / b
# saida

print("------------------ OPERADORES ARITIMÉTICOS--------------")
print("")
print("-----------------------SOMA------------------------")
print("")
print(f"O resultado da soma de {a} MAIS {b} é igual a : {soma}")
print("")
print("-----------------------SUBTRACAO------------------------")
print("")
print(f"O resultado da soma de {a} MENOS {b} é igual a : {subtraçao}")
print("")
print("------------------------------------------")
print("----------------------MULTIPLICAO------------------------")
print("")
print(f"O resultado da soma de {a} vezes {b} é igual a : {mutiplicaçao}")
print("")
print("------------------------------------------")
print("----------------------DIVISAO------------------------")
print("")
print(f"O resultado da soma de {a} dividido {b} é igual a : {divisao}")
```

```python
# dias da semana

# Entrada
dia_numero = int(input("Digite um número de 1 a 7 (1 = Domingo, 7 = Sábado): "))

# Processamento e Saída
if dia_numero == 1:
    print("O dia correspondente é Domingo.")
elif dia_numero == 2:
    print("O dia correspondente é Segunda-feira.")
elif dia_numero == 3:
    print("O dia correspondente é Terça-feira.")
elif dia_numero == 4:
    print("O dia correspondente é Quarta-feira.")
elif dia_numero == 5:
    print("O dia correspondente é Quinta-feira.")
elif dia_numero == 6:
    print("O dia correspondente é Sexta-feira.")
elif dia_numero == 7:
    print("O dia correspondente é Sábado.")
else:
    print("Número inválido. Por favor, digite um número entre 1 e 7.")
```

```python
# Ano bissexto

# Entrada
ano = int(input("Digite um ano: "))

# Processamento e Saída
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")
```

```python
# Desconto em compras

# Entrada
valor_compra = float(input("Digite o valor da compra: "))

# Processamento do desconto
desconto = 0
if valor_compra >= 100:
    desconto = valor_compra * 0.10 # 10% de desconto
elif valor_compra >= 50 and valor_compra < 100:
    desconto = valor_compra * 0.05 # 5% de desconto
# Caso contrário (valor_compra < 50), o desconto permanece 0

valor_final = valor_compra - desconto

# Saída
print(f"Valor original da compra: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: R$ {desconto:.2f}")
print(f"Valor final a pagar: R$ {valor_final:.2f}")
```
